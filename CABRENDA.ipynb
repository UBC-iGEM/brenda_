```python
import numpy as np
from matplotlib import pyplot as plt
from brendapyrser import BRENDA

dataFile = "/mnt/c/Users/nikbu/OneDrive/Desktop/TreeSAPP519/brenda_2025_1.txt"
brenda = BRENDA(dataFile)

BRENDA_KMs = np.array(
    [v for r in brenda.reactions for v in r.KMvalues.get_values()]
)

values = BRENDA_KMs[(BRENDA_KMs < 1000) & (BRENDA_KMs >= 0)]

plt.hist(values, bins=50)
plt.title(f"Median KM value: {np.median(values)}")
plt.xlabel("Km (mM)")
plt.ylabel("Frequency")
plt.show()
```

```python
ph_raw = [v for r in brenda.reactions for v in r.PH.get_values()]
ph_values = [float(p) for p in ph_raw if isinstance(p, (int, float))]
ph_values = np.array(ph_values)
ph_values = ph_values[(ph_values >= 1) & (ph_values <= 14)]

plt.hist(ph_values, bins=28)
plt.title(f"Median pH optimum: {np.median(ph_values):.2f}")
plt.xlabel("pH")
plt.ylabel("Frequency")
plt.show()

print(f"Minimum and maximum pH optima in database: {ph_values.min()} to {ph_values.max()}")
```
